```python
# Kom√≥rka 1: przygotowanie ≈õrodowiska
!pip -q install pyjwt python-dotenv

import os, time, jwt
from datetime import datetime
from dotenv import load_dotenv

# Colab nie czyta .env - ustawiamy secret w runtime
os.environ["JWT_SECRET"] = os.environ.get("JWT_SECRET", "demo_colab_secret")
SECRET = os.getenv("JWT_SECRET")
ALGO = "HS256"

def make_token(email, ttl=120):
    """Zwraca token JWT dla podanego emailu (ttl w sekundach)."""
    payload = {"sub": email, "exp": int(time.time()) + int(ttl)}
    return jwt.encode(payload, SECRET, algorithm=ALGO)

def decode_token(tok):
    """Zwraca zdekodowany payload tokena lub rzuca wyjƒÖtek je≈õli niepoprawny."""
    return jwt.decode(tok, SECRET, algorithms=[ALGO])
```
